```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temprature')
plt.title('Weather')

plt.plot(x,y, color = 'green', linestyle = 'dotted', linewidth = 5)
```

```python
plt.plot(x,y, 'g+-.')
```

```python
plt.plot(x, y, color = 'green', linestyle = '-.', marker = '+', markersize = 20) # can use hexa decimal for it
```

```python
plt.plot(x,y, ls = '-.', lw = 3)
```
