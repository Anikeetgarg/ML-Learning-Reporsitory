```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits
```

```python
data = load_digits()

dir(data)
```

```python
X = pd.DataFrame(data = data.data, columns= data.feature_names)
y = pd.DataFrame(data = data.target)
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
```

```python
model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params' : {
            'C': [1,10,20],
            'kernel': ['rbf','linear']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'n_estimators': [1,5,10,50,100,200,500 ,1000]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params': {
            'C': [1,5,10]
        }
    },
    'gaussian_bayes' : {
        'model' : GaussianNB(),
        
        'params' : {}  
    },
    'multinomail_bayes': {
        'model': MultinomialNB(),
        'params' : {}
    },
    'DecisionTree' : {
        'model': DecisionTreeClassifier(),
        'params': {}
    }
}
```

```python
score = []

from sklearn.model_selection import GridSearchCV

for model_name, mp in model_params.items():
    clf = GridSearchCV(estimator= mp['model'], param_grid=mp['params'], cv = 5, return_train_score= False)
    clf.fit(X, y)
    score.append({
        'model_name' : model_name,
        'model_best_score' : clf.best_score_ ,
        'model_best_parm' : clf.best_params_
    })
    
x = pd.DataFrame(score, columns = ['model_name', 'model_best_score', 'model_best_parm'])
```
