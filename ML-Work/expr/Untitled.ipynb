```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2,-2,500)
y = x**2
```

```python
plt.plot(x,y)
```

```python
x = np.linspace(-1.4, 1.4, 30)

plt.plot(x, x, 'g--')
plt.plot(x, x**2, 'r:')
plt.plot(x, x**3, 'b^')

plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x, x)
plt.subplot(2,2,2)
plt.plot(x, x**2)
plt.subplot(2,2,3)
plt.plot(x, x**3)
plt.subplot(2,2,4)
plt.plot(x**4)
```

```python
plt.subplot(2,2,1)
plt.plot(x, x)
plt.subplot(2,2,2)
plt.plot(x, x**2)
plt.subplot(2,1,2)
plt.plot(x, x**3)
```

```python
x = np.linspace(-1.5, 1.5, 30)
px = 0.8
py = px**2

plt.plot(x,x**2, 'b-')
plt.plot(px, py, 'ro')

plt.text(px - 1, py, 'Beautiful point', weight = 'heavy')
plt.text(px + 0.05, py - 0.4, 'x = %0.2f\ny = %0.2f'%(px,py), rotation = -30, color = "grey")
plt.text(0,1.5, 'Square Function\n$y = x^2$' , horizontalalignment= "center", fontsize = 20, color = "blue")
```
