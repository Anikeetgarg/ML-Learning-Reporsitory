```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
heart_data = pd.read_csv('heart.csv')
heart_data.head()
```

```python
X = heart_data.drop(['HeartDisease'], axis=1)
y = heart_data['HeartDisease']
```

```python


object_cols = [col for col in heart_data.columns if heart_data[col].dtype == 'object']\

ohe_cols = [col for col in object_cols if heart_data[col].unique().size < 10]
```

```python
# preprocessing

from sklearn.preprocessing import OneHotEncoder

oh = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

transformend_data = oh.fit(X[ohe_cols])

ohe_transformed_col_names = transformend_data.get_feature_names_out()

heart_processed_data = pd.DataFrame(oh.transform(X[ohe_cols]), columns=ohe_transformed_col_names)

X.drop(ohe_cols, axis= 1, inplace=True)

X.head()
```

```python
all_data = pd.concat([X, heart_processed_data], axis=1)
all_data.head()
```

```python

heart_data.describe()
```

```python
features_to_scate = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']

from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()

all_data[features_to_scate] = mms.fit_transform(all_data[features_to_scate])
```

```python
all_data.describe()
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier

model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params' : {
            'C': [1,10,20],
            'kernel': ['rbf','linear']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'n_estimators': [1,5,10,50,100]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params': {
            'C': [1,5,10]
        }
    }
}
```

```python
from sklearn.model_selection import GridSearchCV

scores = []

for model_name, mp in model_params.items():
    cfl = GridSearchCV(estimator= mp['model'], param_grid=mp['params'], cv = 5, return_train_score=False)
    
    cfl.fit(all_data, y)
    
    scores.append({
        'model_name' : model_name,
        'model_best_score' : cfl.best_score_,
        'model_best_param' : cfl.best_params_
    })
    
p = pd.DataFrame(data = scores, columns=['model_name', 'model_best_score', 'model_best_param'])
```
